```python
#Importo pandas e il file in formato csv.
import pandas as pd
import numpy as np
df=pd.read_csv("C:\\Users\\gabri\\Desktop\\PERCORSO EPICODE\\M3_PYTON\\PRATICA\\ESERCIZIO FINALE\\owid-covid-data.csv")
```

```python
#Exploratory Data Analysis - EDA
#Prima overview dei dati:
display(df.head())
#Visualizzazione delle informazioni generali sul dataset: 
df.info()
#Analisi delle colonne numeriche: 
df.describe()
```

```python
#Analisi delle colonne categoriche:
for colonna in df.columns:
    if df[colonna].dtype == 'object':
        print(f"\nColonne: con dati categorici {colonna}")
        print(f"Tipo:", df[colonna].dtype)
        print(f"Valori unici:", df[colonna].unique())
        print(f"Conteggio Valori unici:", df[colonna].nunique())
        print(f"Frequenze di volte che compare ciascun valore:")
        print(df[colonna].value_counts())
```

```python
#Individuazione dei dati nulli
for colonna in df.columns:
    conteggio_nulli= df[colonna].isna().sum()
    percentuale_nulli= round((conteggio_nulli/df.shape[0])*100,2)
    print(f"{colonna} contiene: somma valori nulli: {conteggio_nulli}  e % valori nulli: {percentuale_nulli}")
```

```python
#DOAMNDA 2 a 
continenti= df[df["continent"].notna()]
ultima_data= continenti.sort_values("date").groupby("location").tail(1)
casi_per_continente= ultima_data.groupby("continent")["total_cases"].sum()
casi_per_continente
```

```python
#DOAMNDA 2 b
totale_mondiale = casi_per_continente.sum()
percentuali_mondiali = (casi_per_continente / totale_mondiale) * 100
print(round(percentuali_mondiali,2))
```

```python
#DOAMNDA 3

import seaborn as sns
df["date"] = pd.to_datetime(df["date"])
italia = df[df["location"] == "Italy"]
italia_2022 = italia[italia["date"].dt.year == 2022]
italia_2022 = italia_2022[italia_2022["new_cases"].notna() & (italia_2022["new_cases"] > 0)]
#RISPOSTA a
print(f"Evoluzione dei casi totali in Italia nel 2022")
sns.lineplot(data=italia_2022, x="date", y="total_cases")
```

```python
#RISPOSTA b
print(f"Nuovi casi giornalieri registrati in Italia nel 2022")
sns.lineplot(data=italia_2022, x="date", y="new_cases")
```

```python
#DOAMNDA 4
import seaborn as sns
df["date"] = pd.to_datetime(df["date"])
paesi = ["Italy", "Germany", "France"]
terapia_intensiva  = df[df["location"].isin(paesi)]
inizio = "2022-05-01"
fine = "2023-04-30"
terapia_intensiva  = terapia_intensiva [(terapia_intensiva ["date"] >= inizio) & (terapia_intensiva ["date"] <= fine)]
terapia_intensiva  = terapia_intensiva [terapia_intensiva ["icu_patients"].notna()]

#RISPOSTA a
print(f"Boxplot: pazienti in terapia intensiva (ICU) da Maggio 2022 a Aprile 2023")
sns.boxplot(data=terapia_intensiva , x="location", y="icu_patients")
```

```python
#RISPOSTA b
print(f"In base al boxplot, la Germania mostra una mediana e una variabilità più alte nel numero di pazienti in terapia intensiva rispetto a\nItalia e Francia. Ciò indica una pressione maggiore sul sistema ospedaliero nel periodo considerato.")
```

```python
#DOAMNDA 5  
import seaborn as sns

df["date"] = pd.to_datetime(df["date"])
paesi = ["Italy", "Germany", "France", "Spain"]
pazienti_ospedalizzati = df[df["location"].isin(paesi)]
pazienti_ospedalizzati_2021 = pazienti_ospedalizzati[pazienti_ospedalizzati["date"].dt.year == 2021]
conta_valori_nulli = pazienti_ospedalizzati_2021["hosp_patients"].isna().sum()

#RISPOSTA a
somma_pazienti_ospedalizzati_per_paese = pazienti_ospedalizzati_2021.groupby("location")["hosp_patients"].sum()
print(f"\nSomma dei pazienti ospedalizzati nel 2021:")
print(somma_pazienti_ospedalizzati_per_paese )
#grafico
sns.barplot(x=somma_pazienti_ospedalizzati_per_paese.index, y=somma_pazienti_ospedalizzati_per_paese.values)
```
